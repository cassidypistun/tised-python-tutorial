```python
import pandas as pd          # pandas  — for working with data tables
import matplotlib.pyplot as plt  # matplotlib — for making charts and graphs

# If these run without errors, you're ready to go.
print("Libraries loaded successfully ✓")
```

## Section 2: A Data Table (DataFrame)

In pandas, a table of data is called a **DataFrame**. Your course labs use DataFrames constantly — Brightway and Pymrio both return results as DataFrames.

The cell below creates a small DataFrame of CO₂ emissions by country. **Run it.**


```python
# A DataFrame is like a spreadsheet table — rows and columns of data

emissions_data = {
    "Country":    ["Canada", "Germany", "France", "China", "Brazil"],
    "Year":       [2021,     2021,      2021,     2021,    2021    ],
    "CO2_Mt":     [571,      675,       306,      11470,   480     ],   # megatonnes CO₂
    "Population": [38.2,     83.2,      67.4,     1412.0,  214.3   ],   # millions
}

df = pd.DataFrame(emissions_data)

# Display the table
df
```

The table appeared below the cell. Each column is a variable, each row is one country.

**Now run this cell** to see a useful derived column — emissions per person:


```python
# Calculate a new column: CO₂ per capita (tonnes per person)
df["CO2_per_capita"] = (df["CO2_Mt"] * 1_000_000) / (df["Population"] * 1_000_000)
df["CO2_per_capita"] = df["CO2_per_capita"].round(1)

# Display the updated table
df
```

## Section 3: Reading Output from a Cell

Your labs will often print results like this. **Run the cell and read the output carefully.**


```python
# Summary statistics
print("═" * 40)
print("  Emissions Dataset Summary")
print("═" * 40)
print(f"  Countries in dataset:   {len(df)}")
print(f"  Highest emitter:        {df.loc[df['CO2_Mt'].idxmax(), 'Country']}")
print(f"  Lowest emitter:         {df.loc[df['CO2_Mt'].idxmin(), 'Country']}")
print(f"  Highest per capita:     {df.loc[df['CO2_per_capita'].idxmax(), 'Country']}")
print(f"  Lowest per capita:      {df.loc[df['CO2_per_capita'].idxmin(), 'Country']}")
print("═" * 40)
```

## Section 4: Your First Chart

This is what your course labs look like when they produce a chart. The code below makes a bar chart of CO₂ emissions by country.

**Run it first to see the default chart.**


```python
# Bar chart — total CO₂ emissions by country

fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(df["Country"], df["CO2_Mt"], color="steelblue", edgecolor="white", linewidth=0.8)

ax.set_title("Total CO₂ Emissions by Country (2021)", fontsize=13, fontweight="bold")
ax.set_xlabel("Country")
ax.set_ylabel("CO₂ Emissions (Mt)")
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

Now you'll make a small change to switch what is being plotted.

**Your turn:** In the cell below, the chart currently shows `"CO2_Mt"` (total emissions). Change it to `"CO2_per_capita"` to see emissions per person instead. The title is also marked for you to update.

> ✏️ **Change only the two lines marked below.**


```python
# ✏️ CHANGE LINE 1 — switch between "CO2_Mt" and "CO2_per_capita"
column_to_plot = "CO2_Mt"

# ✏️ CHANGE LINE 2 — update the chart title to match
chart_title = "Total CO₂ Emissions by Country (2021)"

# --- do not change anything below this line ---
fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(df["Country"], df[column_to_plot], color="darkcyan", edgecolor="white", linewidth=0.8)

ax.set_title(chart_title, fontsize=13, fontweight="bold")
ax.set_xlabel("Country")
ax.set_ylabel(column_to_plot)
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```

## Section 6: The Full Workflow

This final section puts everything together. It mirrors what your course labs look like end-to-end: import → load data → calculate → visualise. 

**Just run each cell in order. No changes needed here.**


```python
# Step 1: Data is already loaded from Section 2
print("Data loaded ✓  —", len(df), "countries,", len(df.columns), "columns")
```

```python
# Step 2: Calculate a ranking column
df["Rank"] = df["CO2_per_capita"].rank(ascending=False).astype(int)
df_sorted = df.sort_values("CO2_per_capita", ascending=True)
print("Ranking calculated ✓")
```

```python
# Step 3: Visualise — horizontal bar chart, sorted by per-capita emissions

fig, ax = plt.subplots(figsize=(8, 5))

colors = ["#2196F3" if c == "Canada" else "#90CAF9" for c in df_sorted["Country"]]

ax.barh(df_sorted["Country"], df_sorted["CO2_per_capita"], color=colors, edgecolor="white")
ax.set_title("CO₂ Emissions Per Capita — Sorted (Canada highlighted)", fontsize=12, fontweight="bold")
ax.set_xlabel("Tonnes CO₂ per person")
ax.grid(axis="x", alpha=0.3)

for i, (val, country) in enumerate(zip(df_sorted["CO2_per_capita"], df_sorted["Country"])):
    ax.text(val + 0.1, i, f"{val:.1f} t", va="center", fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Step 4: Print a summary table
print("
Final results table:")
print(df[["Country", "CO2_Mt", "CO2_per_capita", "Rank"]].to_string(index=False))
```
