### How this notebook works

A notebook is made up of **cells**. There are two kinds:

| Cell type | What it looks like | What it does |
|---|---|---|
| **Markdown** | White background, formatted text | Displays instructions and explanations (like this cell) |
| **Code** | Grey background, coloured text | Contains Python code that the computer will run |

### ▶️ How to run a code cell
Click inside any grey cell below, then press **Shift + Enter** on your keyboard.

The computer will execute the code and show the result directly beneath the cell. A number will appear in the brackets on the left, e.g. `[1]` — this is the order in which cells were run.

## Section 2: Cell Order Matters

Cells must be run **from top to bottom**. Each cell can use information created by earlier cells. Running them out of order causes errors.

**Run these two cells in order (top first, then bottom):**


```python
# Cell A — run this first
city = "Montreal"
temperature_c = -12.5
```

```python
# Cell B — run this second
# It uses the variables defined in Cell A above

print("City:", city)
print("Temperature:", temperature_c, "°C")
```

✅ It worked because you ran Cell A first.

Now try this: **skip ahead and run only the cell below** (Cell D) *without* running Cell C first. What happens?


```python
# Cell C — defines a new variable (DON'T run this one first)
co2_ppm = 421
```

```python
# Cell D — try running this without running Cell C
# You should see a NameError

print("Current CO₂ concentration:", co2_ppm, "ppm")
```

**You should have seen a `NameError`** — the computer didn't know what `co2_ppm` was because Cell C hadn't been run yet.

> 🔧 **Fix it:** Go back and run Cell C, then run Cell D again. The error will disappear.

This is the most common mistake in Jupyter notebooks. In your course labs, always run cells **from top to bottom**.

## Section 3: The Kernel

The **kernel** is the engine that runs your code. Think of it as the computer's memory for this notebook session. It remembers every variable you've defined by running cells.

Look at the top-right corner of this page — you'll see a circle indicator:
- ⭕ **Empty** — the kernel is idle and ready
- ⌛ **Filled / spinning** — the kernel is busy running code

**Run the cell below** — it will take about 3 seconds on purpose so you can watch the indicator.


```python
import time

print("Starting a slow calculation...")
time.sleep(3)
print("Done! The kernel was busy for 3 seconds.")
```

## Section 4: Restarting the Kernel

If something goes wrong in a lab (cells run out of order, strange errors), the fix is almost always:

> **Runtime → Restart session and run all** *(in Google Colab)*
> or
> **Kernel → Restart Kernel and Run All Cells** *(in Jupyter Lab)*

This clears the kernel's memory and re-runs every cell from top to bottom — a clean slate.

**Try it now:**
- In the menu at the top, click **Runtime** → **Restart session and run all**
- Watch all the cells run automatically in order
