## Section 1: What is a Variable?

A **variable** is a named container that stores a value. You create one by writing:

```
name = value
```

The `=` sign means *"store this value under this name"* — it's not a mathematical equals sign.

**Run the cell below:**


````python
# These are variables — each stores a piece of information

country      = "Canada"           # a text value (called a "string")
year         = 2024               # a whole number (called an "integer")
co2_ppm      = 421.5              # a decimal number (called a "float")
above_target = True               # True or False (called a "boolean")

# Print all of them
print("Country:       ", country)
print("Year:          ", year)
print("CO₂ (ppm):     ", co2_ppm)
print("Above target:  ", above_target)
````

## Section 2: Changing a Variable

Now you'll make your first edit. The cell below calculates the temperature in Fahrenheit from a Celsius value.

**Your turn:** Change the value of `temp_celsius` to any temperature you like (try `-20` for a Montreal winter, or `35` for a heat wave). Then press **Shift + Enter** and observe how the output changes automatically.

> ✏️ **Change only the line marked with the comment below.**


````python
# ✏️ CHANGE THIS LINE — try different temperatures
temp_celsius = 22.0

# --- do not change anything below this line ---
temp_fahrenheit = (temp_celsius * 9/5) + 32

print(f"Temperature: {temp_celsius} °C  =  {temp_fahrenheit:.1f} °F")
````

## Section 3: Strings (Text Values)

Text values are called **strings**. They must always be wrapped in quotes (`"like this"` or `'like this'`).

**Run the cell, then try changing the city name on the marked line.**


````python
# ✏️ CHANGE THIS LINE — try a different city name (keep the quotes!)
city = "Montreal"

# --- do not change anything below this line ---
print(f"Selected city: {city}")
print(f"Analysing environmental data for {city}...")
print(f"Report generated for: {city.upper()}")
````

---
## Section 4: Lists — Storing Multiple Values

A **list** stores multiple values in one variable, using square brackets `[ ]`. Lists are used constantly in data analysis — for example, a list of countries, years, or emission values.

**Run the cell below first. Then try adding or removing a country from the list on the marked line.**


````python
# ✏️ CHANGE THIS LINE — add or remove a country (keep the quotes and commas!)
countries = ["Canada", "Germany", "China", "Brazil"]

# --- do not change anything below this line ---
print("Number of countries:", len(countries))
print("Countries being analysed:")

for country in countries:
    print("  -", country)
````

**What's happening in the loop?** The `for` loop goes through each item in your list and runs the `print()` line once per item. Your labs use loops like this to process multiple datasets automatically — you don't need to write them, just understand that they repeat.

## Section 5: Comments — The # Symbol

Comments (lines starting with `#`) are notes written in the code for humans to read. The computer completely ignores them.

In your labs, comments are used to:
- Explain what a line of code does
- Mark which lines you are allowed to change
- Warn you not to change something

**Run the cell and observe — the commented-out line does nothing:**


````python
emission_factor = 0.42   # kg CO₂ per kWh (Canadian grid average)

# emission_factor = 0.82   # ← this line is "commented out" — the computer ignores it
# Uncommenting it (removing the #) would switch to a different value

energy_kwh = 1000

total_emissions = energy_kwh * emission_factor
print(f"Total emissions: {total_emissions:.1f} kg CO₂")
print(f"Emission factor used: {emission_factor} kg CO₂/kWh")
````

**Try it:** Remove the `#` at the start of the second `emission_factor` line (line 3) and run the cell again. The calculation will switch to the new value. Then put the `#` back and run again.

This is called **commenting/uncommenting** a line — a common way labs let you toggle between options.

## Section 6: Putting It Together

This section combines everything above. You'll change a few values and see how the whole output updates.

**Instructions are marked in the code:**


````python
# ══════════════════════════════════════════
#   SIMPLE CARBON FOOTPRINT CALCULATOR
# ══════════════════════════════════════════

# ✏️ CHANGE THESE LINES — fill in your own values
city            = "Montreal"       # your city name
annual_kwh      = 3500             # household electricity use per year (kWh)
grid_factor     = 0.030            # kg CO₂ per kWh (Quebec grid is very clean!)
# grid_factor   = 0.420            # ← uncomment this line to use the Canadian average instead

# --- do not change anything below this line ---
annual_emissions_kg = annual_kwh * grid_factor
annual_emissions_t  = annual_emissions_kg / 1000

print("=" * 45)
print(f"  Carbon Footprint Report — {city}")
print("=" * 45)
print(f"  Annual electricity use:  {annual_kwh:,} kWh")
print(f"  Grid emission factor:    {grid_factor} kg CO₂/kWh")
print(f"  Annual CO₂ emissions:    {annual_emissions_kg:.1f} kg")
print(f"                           ({annual_emissions_t:.3f} tonnes CO₂)")
print("=" * 45)
````
